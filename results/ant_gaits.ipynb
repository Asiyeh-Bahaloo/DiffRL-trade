```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
sns.set()
```

```python
from matplotlib.ticker import MaxNLocator

df = pd.read_csv("data/ant_gaits.csv")

fig, ax = plt.subplots(figsize=(5,3))
sns.barplot(df, x="h", y="gait", ax=ax, color="tab:blue")
ax2 = ax.twinx()
sns.violinplot(df, x="h", y="reward", ax=ax2, grid=False, outline=False, linewidth=0, color="tab:red")

ax.set_ylim((15, 31))
ax2.set_ylim((0, 8700))

ax.set_xlabel("H")
ax.set_ylabel("Gait period (timesteps)")
ax2.set_ylabel("Episode reward")
ax.grid(False)
ax2.grid(False)
ax.yaxis.label.set_color('tab:blue')
# ax.yaxis.ticks.set_color('tab:blue')
ax.tick_params(axis='y', colors='tab:blue')
ax2.yaxis.label.set_color('tab:red')
# ax2.yaxis.ticks.set_color('tab:red')
ax2.tick_params(axis='y', colors='tab:red')

ticker = matplotlib.ticker.EngFormatter(unit='')
ax2.yaxis.set_major_formatter(ticker)
# ax.yaxis.set_major_formatter(ticker)
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.tight_layout()
plt.savefig("../figures/ant_gait_analysis.pdf", bbox_inches = 'tight',pad_inches = 0)
```
